## Model Building

## K-Nearest Neighbors

```python
test_scores = []
train_scores = []
for i in range(1,15):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
```

```python
max_train_score = max(train_scores)
train_scores_index = [i for i, v in enumerate(train_scores) if v== max_train_score]
print('Max train score {} % and k = {}'.format(max_train_score*100,list(map(lambda x : x+1,train_scores_index))))
```

```python
max_test_score = max(test_scores)
test_scores_index = [i for i, v in enumerate(test_scores) if v==max_test_score]
print('Max test score {} % and  k ={}'.format(max_test_score*100,list(map(lambda x:x+1,test_scores_index))))
```

```python
plt.figure(figsize=(12,15))
p = sns.lineplot(range(1,15),train_scores,marker="*",label = 'Train_score')
p = sns.lineplot(range(1,15),test_scores,marker = "o",label = 'Test_score')
```

Highest score of test coming at k=8 so we will take 8 for our final model

```python
knn = KNeighborsClassifier(8)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
#Plotting the decision boundary
value = 20000
width = 20000
plot_decision_regions(X.values,y.values,clf =knn,legend = 2,
                     filler_feature_values = {2:value,3:value,4:value,5:value,6:value,7:value},
                     filler_feature_ranges = {2:value,3:value,4:value,5:value,6:value,7:value},
                     X_highlight = X_test.values)
plt.title('KNN with Diabetes Data')
plt.show()
```

## Confusion matrix

```python
y_prd = knn.predict(X_test)
```

```python
cm = metrics.confusion_matrix(y_test,y_prd)
p = sns.heatmap(pd.DataFrame(cm),annot=True,cmap = 'YlGnBu',fmt ='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
#Classification report
print(classification_report(y_test,y_prd))
```

## ROC-AUC CURVE

```python
y_pred_proba = knn.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```
